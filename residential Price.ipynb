```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import Series, DataFrame
import time
import numpy as np

adds =[]
loadurls = []

with open('suumo.txt','r') as f:
    tokyos = f.readlines()
    for tokyo in tokyos:
        lis = tokyo.split(';')
        adds.append(lis[0].strip())
        loadurls.append(lis[1].strip())

for idx_i,(add,fsurl) in enumerate(zip(adds,loadurls)):
    
    print('start')
    #url情報
    url = fsurl

    #最終ページを取得
    result = requests.get(url)
    c = result.content
    soup = BeautifulSoup(c,"lxml")
    body = soup.find("body")
    pages = body.find_all("div",{'class':'pagination pagination_set-nav'})
    pages = body.find_all('ol',{'class':'pagination-parts'})
    lis = pages[0].find_all('li')[-1]
    lastpage = int(lis.text)

    urls = []
    urls.append(url)
    for pageidx in range(lastpage - 1):
        num = str(pageidx + 2)
        urls.append(url + '&pn=' + num)

    name = [] #マンション名
    address = [] #住所
    locations0 = [] #立地1つ目（最寄駅/徒歩~分）
    locations1 = [] #立地2つ目（最寄駅/徒歩~分）
    locations2 = [] #立地3つ目（最寄駅/徒歩~分）
    age = [] #築年数
    height = [] #建物高さ
    floor = [] #階
    rent = [] #賃料
    admin = [] #管理費
    shikikin = [] #敷金
    reikin = [] #礼金
    floor_plan = [] #間取り
    area = [] #専有面積

    for url in urls:
        #データ取得
        result = requests.get(url)
        c = result.content
        soup = BeautifulSoup(c,'lxml')
        bukkenList = soup.find('div',{'id':'js-bukkenList'})
        cassetteitems = bukkenList.find_all('div',{'class':'cassetteitem'})

        for cassetteitem in cassetteitems:

            tbodies = cassetteitem.find_all('tbody')

            for lp in range(len(tbodies)):
                #name
                name.append(cassetteitem.find('div',{'class':'cassetteitem_content-title'}).text)
                #address
                address.append(cassetteitem.find('li',{'class':'cassetteitem_detail-col1'}).text)

            #location
            loc = cassetteitem.find('li',{'class':'cassetteitem_detail-col2'})
            sublocs = loc.find_all('div',{'class':'cassetteitem_detail-text'})

            for lp in range(len(tbodies)):
                for idx,sta in enumerate(sublocs):

                    if idx == 0:
                        locations0.append(sta.text)
                    elif idx == 1:
                        locations1.append(sta.text)
                    elif idx == 2:
                        locations2.append(sta.text)

            stairs_frombuild = cassetteitem.find('li',{'class':'cassetteitem_detail-col3'})

            for lp in range(len(tbodies)):
                age.append(stairs_frombuild.find_all('div')[0].text)
                height.append(stairs_frombuild.find_all('div')[1].text)

            data = []
            for tbody in tbodies:
                tmp_data = tbody.find('tr').find_all('td')
                floor.append(tmp_data[2].text.strip())
                rent_admin = tmp_data[3].text.strip().split('\n')
                rent.append(rent_admin[0])
                admin.append(rent_admin[1])
                tmp_shiki_rei = tmp_data[4].text.strip().split('\n')
                shikikin.append(tmp_shiki_rei[0])
                reikin.append(tmp_shiki_rei[1])
                floor_area = tmp_data[5].text.strip().split('\n')
                floor_plan.append(floor_area[0])
                area.append(floor_area[1])

        print(url)
        time.sleep(5)

    #各リストをシリーズ化
    name = Series(name)
    address = Series(address)
    locations0 = Series(locations0)
    locations1 = Series(locations1)
    locations2 = Series(locations2)
    age = Series(age)
    height = Series(height)
    floor = Series(floor)
    rent = Series(rent)
    admin = Series(admin)
    shikikin = Series(shikikin)
    reikin = Series(reikin)
    floor_plan = Series(floor_plan)
    area = Series(area)

    #各シリーズをデータフレーム化
    suumo_df = pd.concat([name, address, locations0, locations1, locations2, age, height, floor, rent, admin, shikikin, reikin,floor_plan, area], axis=1)

    #カラム名
    suumo_df.columns=['マンション名','住所','立地1','立地2','立地3','築年数','建物高さ','階','賃料','管理費', '敷金','礼金','間取り','専有面積']

    #csvファイルとして保存
    suumo_df.to_csv(add + '.csv', sep = '\t',encoding='utf-16',index=False)
    
    
```
